```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have loaded your data into a DataFrame named 'df'

# Filter the data for the years 2019, 2021, and 2022
years_of_interest = [2019, 2021, 2022]
df_filtered = concat_df[concat_df['Fiscal Year'].isin(years_of_interest)]

# Group the data by 'Fiscal year' and 'Subgroup'
grouped_data = df_filtered.groupby(['Fiscal Year', 'Subgroup']).agg({
    'Percent Proficiency Level 1': 'mean',
    'Percent Proficiency Level 2': 'mean',
    'Percent Proficiency Level 3': 'mean',
    'Percent Proficiency Level 4': 'mean'
}).reset_index()

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Define bar width
bar_width = 0.2

# Define position of bars on X-axis
x = range(len(years_of_interest))

# Plot bars
for i, (label, color) in enumerate([('Proficiency Level 1', 'blue'), ('Proficiency Level 2', 'green'), ('Proficiency Level 3', 'orange'), ('Proficiency Level 4', 'red')]):
    male_values = grouped_data[grouped_data['Subgroup'] == 'Male']['Percent Proficiency Level '+str(i+1)]
    female_values = grouped_data[grouped_data['Subgroup'] == 'Female']['Percent Proficiency Level '+str(i+1)]

    plt.bar([pos + i * bar_width for pos in x], male_values, width=bar_width, label='Male '+label, color=color)
    plt.bar([pos + i * bar_width + bar_width for pos in x], female_values, width=bar_width, label='Female '+label, color=color, alpha=0.7)

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Percentage Proficiency')
plt.title('Percentage Proficiency Levels by Year and Gender')
plt.xticks([pos + 1.5 * bar_width for pos in x], years_of_interest)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with the merged data

# Calculate average proficiency levels for each year
avg_2019 = concat_df[concat_df['Fiscal Year'] == 2019][['Percent Proficiency Level 1', 'Percent Proficiency Level 2', 'Percent Proficiency Level 3', 'Percent Proficiency Level 4']].mean()
avg_2021 = concat_df[concat_df['Fiscal Year'] == 2021][['Percent Proficiency Level 1', 'Percent Proficiency Level 2', 'Percent Proficiency Level 3', 'Percent Proficiency Level 4']].mean()
avg_2022 = concat_df[concat_df['Fiscal Year'] == 2022][['Percent Proficiency Level 1', 'Percent Proficiency Level 2', 'Percent Proficiency Level 3', 'Percent Proficiency Level 4']].mean()

# Combine the averages into a single DataFrame
avg_df = pd.DataFrame({
    '2019': avg_2019,
    '2021': avg_2021,
    '2022': avg_2022
})

# Create subplots for pie charts
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(25, 10))

years = avg_df.columns

# Generate pie charts for each year
for i, year in enumerate(years):
    ax = axes[i]
    ax.pie(avg_df[year], labels=avg_df.index, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'orange', 'red'])
    ax.set_title(f'Average Proficiency Levels - {year}')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define proficiency levels
proficiency_levels = ['Percent Proficiency Level 1', 'Percent Proficiency Level 2', 'Percent Proficiency Level 3', 'Percent Proficiency Level 4']

# Calculate average proficiency levels for each year
avg_2019 = concat_df[concat_df['Fiscal Year'] == 2019][proficiency_levels].mean()
avg_2021 = concat_df[concat_df['Fiscal Year'] == 2021][proficiency_levels].mean()
avg_2022 = concat_df[concat_df['Fiscal Year'] == 2022][proficiency_levels].mean()

# Combine the averages into a single DataFrame
avg_df = pd.DataFrame({
    '2019': avg_2019,
    '2021': avg_2021,
    '2022': avg_2022
})

# Create subplots for pie charts
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))

years = avg_df.columns

# Generate pie charts for each proficiency level
for i, proficiency_level in enumerate(proficiency_levels):
    ax = axes[i]
    ax.pie(avg_df.loc[proficiency_level], labels=years, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'orange'])
    ax.set_title(f'Average Proficiency Level - {proficiency_level}')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
